### Κλάσεις και αντικείμενα
1.1 Δημιουργήστε μια κλάση με όνομα Interval που αναπαριστά ένα διάστημα της μορφής \[from, to\). Ο κατασκευαστής της κλάσης να δέχεται τιμές για τα δύο όρια και να προκαλεί εξαίρεση ValueError αν το from είναι μεγαλύτερο ή ίσο του to. Η κλάση θα έχει μέθοδο length() που επιστρέφει το μήκος του διαστήματος. Στη συνέχεια, ζητήστε από τον χρήστη να εισάγει 3 διαστήματα και για κάθε διάστημα να εμφανίζεται είτε μήνυμα σφάλματος αν προκύψει εξαίρεση είτε το μήκος του διαστήματος.

**Για προχωρημένους**

Προσθέστε στην κλάση τη μέθοδο overlaps(other) που θα επιστρέφει True αν το διάστημα αλληλεπικαλύπτεται με ένα άλλο διάστημα other, αλλιώς θα επιστρέφει False. Για παράδειγα τα διαστήματα \[1,3\) και \[2,4\) αλληλοεπικαλύπτονται, ενώ τα \[1,3\) και \[3,5\) όχι. Επιπλέον, ελέγξτε αν οποιοδήποτε από τα διαστήματα που εισήχθησαν αλληλεπικαλύπτονται μεταξύ τους και εμφανίστε τα ζεύγη που αυτό συμβαίνει.

````python
# Λύση 1.1
class Interval:
  def __init__(self, a, b):
      if a >= b:
          raise ValueError(f"Μη έγκυρο διάστημα: Πρέπει {a} < {b}.")
      self.a = a
      self.b = b
  def length(self):
      return self.b - self.a
  def overlaps(self, other):
        return max(self.a, other.a) < min(self.b, other.b)


def main():
    while True:
       a1 = input("Δώσε το πρώτο διάστημα Start: ")
       b1 = input("Δώσε το πρώτο διάστημα End: ")
       try:
         a1, b1 = float(a1), float(b1)
         a1_interval = Interval(a1, b1)
         print(f"Μήκος διαστήματος : {a1_interval.length()}")
         break
       except ValueError as e:
         print(f"Σφάλμα: {e}")
         print(f"Ξαναπροσπάθησε")
    while True:
       a2 = input("Δώσε το Δεύτερο διάστημα Start: ")
       b2 = input("Δώσε το Δεύτερο διάστημα End: ")
       try:
         a2, b2 = float(a2), float(b2)
         a2_interval = Interval(a2, b2)
         print(f"Μήκος διαστήματος : {a2_interval.length()}")
         break
       except ValueError as e:
         print(f"Σφάλμα: {e}")
         print(f"Ξαναπροσπάθησε")
    while True:
       a3 = input("Δώσε το Τρίτο διάστημα Start: ")
       b3 = input("Δώσε το Τρίτο διάστημα End: ")
       try:
         a3, b3 = float(a3), float(b3)
         a3_interval = Interval(a3, b3)
         print(f"Μήκος διαστήματος : {a3_interval.length()}")
         break
       except ValueError as e:
         print(f"Σφάλμα: {e}")
         print(f"Ξαναπροσπάθησε")
    if a1_interval.overlaps(a2_interval):
      print(f"Το πρώτο και το Δεύτερο διάστημα Επικαλύπτονται")
    if a2_interval.overlaps(a3_interval):
      print(f"Το πρώτο και το Τρίτο διάστημα Επικαλύπτονται")
    if a2_interval.overlaps(a3_interval):
      print(f"Το Δεύτερο και το Τρίτο διάστημα Επικαλύπτονται")
# Εναλακτικός τρόπος σύγκρισης των ζευγών διαστημάτων
    intervals = [a1_interval, a2_interval, a3_interval]
    labels = ["Πρώτο", "Δεύτερο", "Τρίτο"]
    overlappedIntervals = []  # ζεύγη που επικαλύπτονται
    overlappedLabels    = []  # ζεύγη που επικαλύπτονται Labels
    for i in range(len(intervals)):
      for j in range(i+1, len(intervals)):
          if intervals[i].overlaps(intervals[j]):
              overlappedIntervals.append((intervals[i], intervals[j]))
              overlappedLabels.append((labels[i], labels[j]))
# Εμφάνιση των αποτελεσμάτων
    if overlappedLabels:
      print("\nΕπικαλυπτόμενα διαστήματα:")
      for i, (l1, l2) in enumerate(overlappedLabels):
          print(
              f"Το {l1} ({overlappedIntervals[i][0].a}, {overlappedIntervals[i][0].b}) "
              f" και το {l2} ({overlappedIntervals[i][1].a}, {overlappedIntervals[i][1].b}) "
              " διάστημα επικαλύπτονται"
          )
    else:
        print("\nΔεν υπάρχουν επικαλυπτόμενα διαστήματα.")

if __name__ == "__main__":
    main()
````

### Αρχεία
1.3 Κατεβάστε το αρχείο bigmac.csv από το <https://calmcode.io/datasets>. Διαβάστε τα δεδομένα με το module csv της Python. Εντοπίστε τις 3 χώρες με τις υψηλότερες τιμές big mac σε τιμές δολαρίων Αμερικής (dollar_price).

````python
!wget https://calmcode.io/static/data/bigmac.csv -O bigmac.csv
!ls -l
!pwd
! head bigmac.csv
````

````python
# Λύση 1.3
import csv

countries = {}

with open("bigmac.csv", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        try:
            country = row["name"]               # χώρα
            price = float(row["dollar_price"])  # τιμή σε δολάρια
            if country not in countries or price > countries[country]:
                countries[country] = price
        except ValueError:
            continue
cnt_countries = 0

# αρχικοποίηση top 3
max1, max2, max3 = 0, 0, 0
c1, c2, c3 = "", "", ""

for j, cuntr in enumerate(countries):
    cnt_countries += 1
    price = countries[cuntr]

    # ενημέρωση top 3
    if price > max1:
        max3, c3 = max2, c2
        max2, c2 = max1, c1
        max1, c1 = price, cuntr
    elif price > max2:
        max3, c3 = max2, c2
        max2, c2 = price, cuntr
    elif price > max3:
        max3, c3 = price, cuntr

    # εκτύπωση με σπασίματα κάθε 5
    print(f"({j+1}). {cuntr}", end=" ")
    if (j+1) % 5 == 0:
        print("\n")

print(f"\nΔιάβασα {cnt_countries} χώρες")

print("\nΟι 3 μεγαλύτερες τιμές:")
print(f"1. {c1}: {max1:.2f} USD")
print(f"2. {c2}: {max2:.2f} USD")
print(f"3. {c3}: {max3:.2f} USD")



# Εναλλακτικός τρόπος με Ταξινόμηση λίστας
max_prices = sorted(countries.items(), key=lambda x: x[1], reverse=True)

print("Τις 3 χώρες με τις υψηλότερες τιμές Big Mac:")
for country, price in max_prices[:3]:
    print(f"{country}: {price:.2f} USD")
````

1.4 Γράψτε κώδικα που να συνδέεται στη βάση δεδομένων Chinook που θα κατεβάσετε τοπικά από το <https://www.sqlitetutorial.net/sqlite-sample-database/>, χρησιμοποιώντας τη βιβλιοθήκη sqlite3. Στη συνέχεια να θέτει ερώτημα προς τη βάση που να επιστρέφει τα ονόματα και τα επώνυμα όλων των πελατών που υπάρχουν στον πίνακα Customers.

````python
!wget https://www.sqlitetutorial.net/wp-content/uploads/2018/03/chinook.zip -O chinook.zip
!unzip -o chinook.zip
!ls -l
````

````python
# Λύση 1.4
import sqlite3

# Σύνδεση στη βάση
conn = sqlite3.connect("chinook.db")

# 1ος cursor: όλα τα πεδία του πίνακα Customers
cur_all = conn.cursor()
cur_all.execute("SELECT * FROM Customers;")
rows_all = cur_all.fetchall()

col_names = [desc[0] for desc in cur_all.description]
print("Όλα τα πεδία από Customers:")
print(col_names, "\n")
for row in rows_all[:5]:  # 5 πρώτες εγγραφές
    print(row)

print("\nΣύνολο εγγραφών:", len(rows_all))

# 2ος cursor: μόνο FirstName και LastName
cur_names = conn.cursor()
cur_names.execute("SELECT FirstName as 'ONOMA', LastName as 'ΕΠΙΘΕΤΟ' FROM Customers order by LastName,FirstName;")
rows_names = cur_names.fetchall()

col_names = [desc[0] for desc in cur_names.description]
print("Πεδία από Customers:")
print(col_names, "\n")

print("\nΌνομα – Επώνυμο πελατών:")
for i,(first, last) in enumerate(rows_names):
    print(f"{(i+1):3d}.  {last} {first}")
print(f"\n\n Διάβασα {i+1} Πελάτες στη βάση")
# Κλείσιμο σύνδεσης
conn.close()
````
