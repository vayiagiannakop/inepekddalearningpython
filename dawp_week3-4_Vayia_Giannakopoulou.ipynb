```python
import numpy as np
from tabulate import tabulate

#------------------------------------------------------------------------
#  1) Φτιάξε έναν 1D πίνακα ids = [101, 102, 103, 104, 105] με dtype=int32.
#     Πως φτιάχνετε; ids = np.array([101,102,103,104,105], dtype=np.int32)
#------------------------------------------------------------------------

ids = np.array([101, 102, 103, 104, 105], dtype=np.int16)
print(f"O 1D πίνακας \n {ids}, \n Με το shape ids.shape {ids.shape} \n")

idsvector1=ids.reshape(len(ids),1)
print(f"O 1D πίνακας ΣΤΗΛΗ \n {idsvector1}, \n Με το shape idsvector1.shape {idsvector1.shape}\n")
print(tabulate(idsvector1, tablefmt="double_grid"),"\n")

idsvector2=ids.reshape(1,len(ids))
print(f"O 1D πίνακας ΓΡΑΜΜΗ \n {idsvector2}, \n Με το shape idsvector2.shape {idsvector2.shape}\n")
print(tabulate(idsvector2, tablefmt="double_grid"),"\n")

#------------------------------------------------------------------------
# 2) Φτιάξε έναν 2D πίνακα base_people για 5 άτομα με στήλες
#    [ύψος_cm, βάρος_kg].   Δώσε δικές σου τιμές, π.χ.
#------------------------------------------------------------------------

[[185, 75], [170, 68], [192, 82], [178, 70], [175, 77]]
base_people = np.array([
    [185, 75],
    [170, 68],
    [192, 82],
    [178, 70],
    [175, 77],
], dtype=np.int32)
print(f"O 2D πίνακας base_people \n {base_people}, \n Με το shape base_people.shape {base_people.shape}\n")

print(tabulate(base_people, tablefmt="double_grid"),"\n")

#Ειδικός μετασχηματισμός για να τυπώσουμε όμορφα τον πίνακα με tabulate class
headers_base_people = ["id", "ύψος_cm", "βάρος_kg"]
# κολλάμε την στήλη ids μπροστά από το base_people
base_people_with_ids = np.column_stack([ids, base_people])
print(tabulate(base_people_with_ids, headers=headers_base_people, tablefmt="double_grid"))

#------------------------------------------------------------------------
# 3)  Φτιάξε έναν 3D πίνακα groups σχήματος (3, 5, 2) όπου ο άξονας-0
#     είναι η ομάδα (0..2). Χρησιμοποίησε παραλλαγές του base_people
#     (π.χ. πρόσθεσε/αφαίρεσε λίγα εκατοστά/κιλά με broadcasting).
#------------------------------------------------------------------------

offsets = np.array([
     [ 0,  0],
     [-2, +2],
     [+3, -2],
 ], dtype=np.int16)
print(f"O 2D πίνακας offsets \n {offsets}, \n Με το shape offsets.shape {offsets.shape} \n")
print(tabulate(offsets, tablefmt="double_grid"),"\n")
offsets=offsets.reshape(3,1,2)
print(f"O 2D πίνακας offsets με reshape(3,1,2) \n {offsets}, \n Με το shape offsets.shape {offsets.shape} \n")

# Τυπώνουμε τα offsets για κάθε ομάδα
for i, group in enumerate(offsets):
    print(f"Offsets για την Ομάδα {i}")
    print(tabulate(group, tablefmt="double_grid"), "\n")

# O groups πίνακας γίνεται διάστασης (1,5,2)
groups = base_people[None, :, :]
print(f"groups = base_people[None, :, :]\n {groups}, \n Με το shape groups.shape {groups.shape} \n")
# Τώρα το Broadcasting
# O groups πίνακας διάστασης (1,5,2)
# O πίνακας offsets διάστασης (3,1,2)
# Το broadcasting είναι εφικτό επειδή οι διαστάσεις
#  (3,1,2) και (1,5,2) είναι συμβατές
#  και "επεκτείνονται" αυτόματα σε κοινό σχήμα (3,5,2).
groups=groups+offsets
print(f"groups= groups + offsets \n {groups}, \n Με το shape groups.shape {groups.shape} \n")
# Θα το κάνω τώρα χωρίς επανάλληψη
# for i, group in enumerate(groups):
#     print(f"Ομάδα {i}")
#     print(tabulate(group, tablefmt="double_grid"), "\n")

print("Ομάδα 0")
print(tabulate(groups[0], tablefmt="double_grid"), "\n")

print("Ομάδα 1")
print(tabulate(groups[1], tablefmt="double_grid"), "\n")

print("Ομάδα 2")
print(tabulate(groups[2], tablefmt="double_grid"), "\n")

#------------------------------------------------------------------------
# 4) Δημιούργησε με np.arange και np.linspace μικρά βοηθητικά διανύσματα.
#------------------------------------------------------------------------

v1 = np.arange(0, 10, 2, dtype=np.int32)
print(f"v1 = np.arange(0, 10, 2, dtype=np.int32)\n {v1}, \n Με το shape groups.shape {v1.shape} \n")
print(tabulate(v1.reshape(1,len(v1)), tablefmt="double_grid"),"\n")
v2 = np.linspace(0.0, 1.0, num=5)
print(f"v2 = np.linspace(0.0, 1.0, num=5)\n {v2}, \n Με το shape groups.shape {v2.shape} \n")
print(tabulate(v2.reshape(1,len(v2)), tablefmt="double_grid"),"\n")

#------------------------------------------------------------------------
# 5) Φτιάξε np.zeros((5,2)), np.ones((2,2)),
#           np.eye(3) και np.full((5,), fill_value=100).
#------------------------------------------------------------------------

Z = np.zeros((5, 2))
print(f"Z = np.zeros((5, 2))\n {Z}, \n Με το shape groups.shape {Z.shape} \n")
print(tabulate(Z, tablefmt="double_grid"),"\n")

O = np.ones((2, 2), dtype=np.int8)
print(f"O = np.ones((2, 2), dtype=np.int8)\n {O}, \n Με το shape groups.shape {O.shape} \n")
print(tabulate(O, tablefmt="double_grid"),"\n")

I = np.eye(3, dtype=np.int8)
print(f"I = np.eye(3, dtype=np.int8)\n {I}, \n Με το shape groups.shape {I.shape} \n")
print(tabulate(I, tablefmt="double_grid"),"\n")

F = np.full((5,), fill_value=100, dtype=np.int32)
print(f"F = np.full((5,), fill_value=100, dtype=np.int32)\n {F}, \n Με το shape groups.shape {F.shape} \n")
print(tabulate(F.reshape(1,len(F)), tablefmt="double_grid"),"\n")
```

## 6) Συνένωση (stack) & Διαίρεση (split)

2. Κόψε τον `all_people` σε 3 ίσα μέρη των `(5, 2)` με `np.split`.

3. Δείξε `np.concatenate` οριζόντια και κατακόρυφα (π.χ. πρόσθεσε μια στήλη/χαρακτηριστικό).



```python
# 1) Στοίβαξε τις 3 ομάδες σε έναν 2D πίνακα (15,2)
# α τροπος με reshape
all_people = groups.reshape(-1, 2)
print("---reshape(-1,2)", all_people)
# β τροπος με vstack
all_people_v = np.vstack([groups[0], groups[1], groups[2]])
print("vstack",all_people_v)
# Σύγκριση των πινάκων με α τροπος και β τρόπο
print("Οι πίνακες που έφταξα με reshape και vstack είναι ίδιοι")
print(np.array_equal(all_people, all_people_v, equal_nan=True))

# ---------------------------------------------------------------
# 2) Κόψε τον all_people σε 3 ίσα μέρη των (5, 2) με np.split.
parts = np.split(all_people, 3, axis=0)    # λίστα 3 στοιχείων, το καθένα (5,2)
# Τώρα θα ξανασυνθέσω την κομμένη λίστα στην again_all_people
again_all_people= np.vstack([parts[0], parts[1], parts[2]])
print("Ο πίνακας που προέκειψε από τα κομματια της λίστας vsplit")
print("που συγχώνευσα ξανα με vstack είναι ίδιος με τον αρχικό πίνακα")
print(np.array_equal(all_people, again_all_people, equal_nan=True))

# ---------------------------------------------------------------
# 3) np.concatenate οριζόντια (axis=1) και κατακόρυφα (axis=0)
p0, p1, p2 = parts
print("\n p0 \n", p0, "\n p1 \n", p1, "\n p2 \n", p2," \n ","_________"*10," \n ")

vert_cat = np.concatenate([p0, p1, p2], axis=0)  # (15,2) — ίδια #στηλών
print("\n np.concatenate([p0, p1, p2], axis=0) \n", vert_cat," \n ","_________"*10," \n ")
print("\n Ο πίνακας που προέκειψε από τα κομματια της λίστας vsplit")
print("p0,p1,p2 που συνένωσα με συγχώνευσα np.concatenate([p0, p1, p2], axis=0) είναι ίδιος με τον αρχικό πίνακα")
print(np.array_equal(all_people, vert_cat, equal_nan=True))

# Φτιάχνω τεσσερεις νεες στήλες ύψος σε cm, βάρος, bmi και height σε m
vert_cat_height_cm= vert_cat[:, 0]
vert_cat_height_cm = vert_cat_height_cm.reshape(-1,1)
vert_cat_height= vert_cat[:, 0] / 100.0
vert_cat_height = vert_cat_height.reshape(-1,1)
vert_cat_weight = vert_cat[:, 1]
vert_cat_weight = vert_cat_weight.reshape(-1,1)
vert_cat_bmi = (vert_cat_weight / (vert_cat_height**2))
vert_cat_bmi = vert_cat_bmi.reshape(-1,1)
print("\n ", vert_cat_bmi," \n ","_________"*10," \n ")
vert_cat_with_bmi = np.concatenate([vert_cat_weight,vert_cat_height_cm,vert_cat_height, vert_cat_bmi], axis=1)
print("\n ", vert_cat_with_bmi," \n ","_________"*10," \n ")
headersvert_bmi = ["Βάρος σε κιλά", "Ύψος σε cm", "Ύψος σε m", "Δείκτης Μάζας Σώματος"]
print(tabulate(vert_cat_with_bmi, headers=headersvert_bmi, tablefmt="rounded_grid", floatfmt=".2f"))

```

2. Βρες τη θέση (`argmax`) του ψηλότερου ατόμου **σε κάθε ομάδα**.



```python
# ==========================================
#  Εκτύπωση των ομάδων
# ==========================================
for i, g in enumerate(groups):
    print(f"Ομάδα {i}")
    print(tabulate(g, headers=['Ύψος(cm)', 'Βάρος(kg)'], tablefmt='rounded_grid'), "\n")
# ==========================================
#  Δείκτες ταξινόμησης ανά ομάδα βάσει ύψους
# ==========================================
sort_idx = np.argsort(groups[:, :, 0], axis=1)
print("\n",groups[:, :, 0], "\n","shape groups[:, :, 0] = ", np.shape(groups[:, :, 0]))
print("Δείκτες ταξινόμησης ανά ομάδα (argsort):")
print("\n",sort_idx, "\n","shape = ", sort_idx.shape)

# =========================================
#  Ταξινομημένες ομάδες (ύψος σε αύξουσα σειρά)
# ==========================================
print("sort_idx[..., None]",sort_idx[..., None],np.shape(sort_idx[...,None]))
sorted_groups = np.take_along_axis(groups, sort_idx[..., None], axis=1)
print("Ταξινομημένες ομάδες κατά ύψος:")
for i, g in enumerate(sorted_groups):
     print(f"Ομάδα {i}")
     print(tabulate(g, headers=['Ύψος(cm)', 'Βάρος(kg)'], tablefmt='rounded_grid'), "\n")

# ==========================================
#  Θέση του ψηλότερου ατόμου (argmax)
# ==========================================
max_idx = np.argmax(groups[:, :, 0], axis=1)
print("Δείκτης (θέση) ψηλότερου ατόμου σε κάθε ομάδα:", max_idx)

# ==========================================
#  Τιμές του ψηλότερου ατόμου ανά ομάδα
# ==========================================
tallest_people = groups[np.arange(groups.shape[0]), max_idx, :]
for i, g in enumerate(tallest_people):
     print(f"Ο ψηλότερος για την Ομάδα {i}")
     print(tabulate(g.reshape(-1,2), headers=['Ύψος(cm)', 'Βάρος(kg)'], tablefmt='rounded_grid'), "\n")

```

## 8) Broadcasting (κεντράρισμα/κανονικοποίηση)

1. Υπολόγισε το **μέσο** ύψος και βάρος.





```python
# Ύψος & βάρος
height = groups[:, :, 0]          # (3,5)
weight = groups[:, :, 1]          # (3,5)

# Μέσοι όροι (ολικά)
mean_h = np.nanmean(height)          # ή np.nanmean(...)
mean_w = np.nanmean(weight)

print(f"Ο μέσος όρος ύψους είναι:  {mean_h:.3f} cm")
print(f"Ο μέσος όρος βάρους είναι:  {mean_w:.3f} κιλά")
```

## 9) Τυχαία δειγματοληψία

1. Δημιούργησε `n=4` τυχαία άτομα επιλέγοντας *γραμμές* από `all_people` με `np.random.choice` χωρίς αντικατάσταση.

2. Υπολόγισε τον μέσο όρο ύψους/βάρους στο τυχαίο δείγμα.



```python
all_people = groups.reshape(-1, 2)
n = 4  # μέγεθος δείγματος
print(np.shape(all_people))
# Επιλογή 4 τυχαίων γραμμών χωρίς επανατοποθέτηση (διαφορετικές)
idx = np.random.choice(all_people.shape[0], size=n, replace=False)
sample = all_people[idx, :]   # οι 4 γραμμές του δείγματος

# Υπολογισμός μέσου όρου ύψους/βάρους στο δείγμα
mean_h = np.nanmean(sample[:, 0])
mean_w = np.nanmean(sample[:, 1])

print("Τυχαίο δείγμα 4 ατόμων:")
print(tabulate(sample, headers=["Ύψος (cm)", "Βάρος (kg)"], tablefmt="rounded_grid"))

print(f"\nΜέσο ύψος δείγματος:  {mean_h:.2f} εκατοστά")
print(f"Μέσο βάρος δείγματος: {mean_w:.2f} κιλά")
```
